## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create bins in which to place values based upon Age
age_bins = [0, 9.90, 14.90, 19.90, 24.9, 29.9, 34.90, 39.90, 9999999]

# Create labels for these bins
bin_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

unique_players["Age Group"] = pd.cut(unique_players["Age"], age_bins, labels=bin_labels)

# Create a GroupBy object based upon "Age Group"
age_group = unique_players.groupby("Age Group")

# Find how many rows fall into each bin
age_count = (age_group["Age"].count())

# Age Percentage
age_perc = age_count / player_count *100

age_analysis_df = pd.DataFrame({"Total Count": age_count, 
                            "Percentage of Players": age_perc})
age_analysis_df
```
