## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spend_item = pd.DataFrame(purchase_df.groupby("SN")["Item ID"].count())
top_spend_price = pd.DataFrame(purchase_df.groupby("SN")["Price"].sum())
top_spend_df = pd.merge(top_spend_item, top_spend_price, on="SN")

top_spend_df.rename(columns={"Item ID":"Purchase Count","Price":"Total Purchase Value"}, inplace=True)

top_spend_df["Average Purchase Price"] = top_spend_df["Total Purchase Value"]/top_spend_df["Purchase Count"]
top_spend_df.sort_values("Total Purchase Value",ascending=False, inplace=True)

top_spend_df["Average Purchase Price"] = top_spend_df["Average Purchase Price"].map("${:.2f}".format)
top_spend_df["Total Purchase Value"] = top_spend_df["Total Purchase Value"].map("${:.2f}".format)

top_spend_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_item = pd.DataFrame(purchase_df.groupby("Item Name")["Item ID"].count())
popular_item_price = pd.DataFrame(purchase_df.groupby("Item Name")["Price"].sum())

popular_items_df = pd.merge(popular_item, popular_item_price, on="Item Name")

popular_items_df.rename(columns = {"Price":"Total Purchase Value"}, inplace=True)

popular_items_df.sort_values("Item ID",ascending=False, inplace=True)
popular_items_df["Item Price"] = popular_items_df["Total Purchase Value"]/popular_items_df["Item ID"]
popular_items_df["Total Purchase Value"]

popular_items_df.rename(columns = {"Item ID":"Item Count"}, inplace=True)

popular_items_df["Total Purchase Value"] = popular_items_df["Total Purchase Value"].map("${:.2f}".format)
popular_items_df["Item Price"] = popular_items_df["Item Price"].map("${:.2f}".format)

popular_items_df.head()
```
